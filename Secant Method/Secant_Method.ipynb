<h1> <center>Znajdowanie ekstremów dowolnej funkcji ciągłej na dwolonym przedziale domkniętym metodą siecznych.</center><h1>

<h2>Algorytm znajdowania ekstremów funkcji g(x) metodą stycznych: </h2>
<h3>1) Różniczkujemy funkcję $g'(x)=f(x)$ </h3>
<h3>2) Wybieramy przedział domknięty$ [a,b]$ taki, że $f(a)\cdot f(b)<0$</h3>
<h3>3)Liczymy punkt przecięcia siecznej z osią OX $$x_{0}=a_{0}-f(a_{0})\frac{b_{0}-a_{0}}{f(b_{0})-f(a_{0})}$$</h3>
<h3>4.a) Jeśli $f(a_{0})f(x_{0})<0$ to nowy przedział jest równy $[a_{1}=a_{0},b_{1}=x_{0}]$  </h3>
<h3>4.b) Jeśli $f(a_{0})f(x_{0})<0$ to nowy przedział jest równy $[a_{1}=x_{0},b_{1}=b_{0}]$ </h3>
<h3>5. Kroki 3,4 powtarzają się aż do ustalonej dokładności  $ \varepsilon$ takiej że $ |a_{n}-b{n}|<\varepsilon $ następnie zwracane jest $x_{0}$ - miejsce zerowe funkcji f(x) czyli ekstermum funkcji g(x). </h3>

### Przykład i ilustracja algorytmu: 
<h2> Szukamy lokalnego maksimum funkcji f(x) na przedziale [-0.5,0.5]
<h3>$$ f(x)= x^3 - x^2 $$<h3>

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-0.5,0.8,100)
y = x**3 - x**2
f = plt.figure()

ax = f.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
f.set_figwidth(10)
f.set_figheight(6)
plt.title('f(x)= x³ - x²')
plt.annotate('Lokalne maksimum', xy=(0, 0), xytext=(0, 0.12),
             arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.plot(x, y)
plt.show()
```

<h2> W pierwszej kolejności liczmy pochodną funkcji f(x), następnie szukamy metodą siecznych miejsca zerowego w danym przedziale. <h2>
<h3>$$ g(x)=f'(x)=  3x^2 - 2x $$<h3>


```python
x = np.linspace(-0.6,0.8,100)
y = 3*x**2 - 2*x
f = plt.figure()
ax = f.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
f.set_figwidth(10)
f.set_figheight(6)
plt.title('g(x)= x³ - x²')
point1 = [-0.5, 1.75]
point2 = [0.5, -0.25 ]

x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]
plt.plot(x, y, 'r',x_values,y_values,'g')
plt.axvline(x=-0.5, color='black', linestyle='--')
plt.axvline(x=0.5, color='black', linestyle='--')
plt.annotate('g (x₁)', xy=(-0.5, 1.76), xytext=(-0.49, 2),
             arrowprops=dict(facecolor='blue', shrink=0.05),
            )
plt.annotate('g (x₂)', xy=(0.5, -0.27 ), xytext=(0.6, -0.40 ),
             arrowprops=dict(facecolor='blue', shrink=0.05),
            )
plt.annotate('x₃', xy=(0.376, 0 ), xytext=(0.36, 0.2 ),
             arrowprops=dict(facecolor='blue', shrink=0.05),
            )
plt.legend(['g(x)','Sieczna','Przedział [-0.5,0.5]'])
plt.show()
```

<h3>Za x₁ oraz x₂ przyjujemy punkty krańcowe przedziału. Punkt przecięcia siecznej z osią OX to x₃.</h3>

```python
x = np.linspace(-0.6,0.8,100)
y = 3*x**2 - 2*x
f = plt.figure()
ax = f.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
f.set_figwidth(10)
f.set_figheight(6)
plt.title('g(x)= x³ - x²')
point1 = [-0.5, 1.75]
point2 = [0.375, -0.3281 ]

x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]
plt.plot(x, y, 'r',x_values,y_values,'g')
plt.axvline(x=-0.5, color='black', linestyle='--')
plt.axvline(x=0.5, color='black', linestyle='--')
plt.annotate('g (x₁)', xy=(-0.5, 1.76), xytext=(-0.49, 2),
             arrowprops=dict(facecolor='blue', shrink=0.05),
            )
plt.annotate('g (x₃)', xy=(0.375, -0.3281), xytext=(0.4, -0.2),
             arrowprops=dict(facecolor='blue', shrink=0.05),
            )
plt.annotate('x₄', xy=(0.2368, 0 ), xytext=(0.25, 0.2 ),
             arrowprops=dict(facecolor='blue', shrink=0.1),
            )
plt.legend(['g(x)','Sieczna','Przedział [-0.5,0.5]'])
plt.show()
```

<h3>Następnie drugą sieczną prowadzimy dla punktu x₁ oraz x₃ znajdujemy punkt przecięcia z osią OX: x₄.</h3>

```python
x = np.linspace(-0.6,0.8,100)
y = 3*x**2 - 2*x
f = plt.figure()
ax = f.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
f.set_figwidth(10)
f.set_figheight(6)
plt.title('g(x)= 3x² - 2x')
point1 = [-0.5, 1.75]
point2 = [0.0122, -0.0239 ]

x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]
plt.plot(x, y, 'r',x_values,y_values,'g')
plt.axvline(x=-0.5, color='black', linestyle='--')
plt.axvline(x=0.5, color='black', linestyle='--')
plt.annotate('g (x₁)', xy=(-0.5, 1.76), xytext=(-0.49, 2),
             arrowprops=dict(facecolor='blue', shrink=0.05),
            )
plt.annotate('g (xᵢ)', xy=(0.0122, -0.0239), xytext=(0.0122, -0.5),
             arrowprops=dict(facecolor='blue', shrink=0.05),
            )
plt.legend(['g(x)','Sieczna','Przedział [-0.5,0.5]'])
plt.show()
```

<h3>Kontunując algorytm i-tą ilość razy otrzymujemy odpowiednie przybliżenie miejsca zerowego pochodnej, czyli ekstema funkcji f(x)<h3>

<h2>Kalkulator maksimów oraz minimów dla zadanej funkcji metodą siecznych</h2>

```python
import math
from sympy import *

def calculate(value, function):
    x = value
    return eval(function)
def zerowe(f,a,b, tolerancja=1e-8):
    fa=calculate(a,f)
    fb=calculate(b,f)
    i=100
    if fa*fb>=0:
        print("W pierwszej iteracji f(a)*f(b) musi być mniejszy niż zero")
        return None
    while(i and math.fabs(a-b)>tolerancja):
        
        if(math.fabs(fa-fb)< tolerancja):
            print("Niewłaściwe punkty początkowe")
            break
        x0= a - fa*(a-b)/(fa-fb)
        f0=calculate(x0,f)
        if fa*f0<0:
            a=a
            b=x0
        elif fb*f0<0:
            b=b
            a=x0
        i=i-1
    if(fa>fb):
        print("Maksimum na zadanym przedziale znajduje się w punkcie: x0 =", round(a - fa*(a-b)/(fa-fb),5))
    if(fa<fb):
        print("Minimum na zadanym przedziale znajduje się w punkcie: x0 = ", round(a - fa*(a-b)/(fa-fb),5))
funkcja= input("f(x)= ")
a= input("Podaj początek przedziału")
b= input("Podaj koniec przedziału")
x=Symbol('x')
funkcja_parse= parse_expr(funkcja)
pochodna = funkcja_parse.diff(x)
print("pochodna wynosi: ",pochodna)
wynik=zerowe(str(pochodna),float(a),float(b))
wynik


```
